```python
import numpy as np
import pandas as pd


df = pd.read_csv("/content/2022 Flight Sales Prediction Model.csv")

df.tail()
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

close = df['Number of Tickets']
ax = close.plot(title='Current sold tickets')
ax.set_xlabel('Date')
ax.set_ylabel('Num of tickets')
plt.show()
```

```python
!pip install AutoTS
from autots import AutoTS
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(df, date_col='Date', value_col='Number of Tickets', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
